```python
import os
from yaml import safe_load

INPUTS_DIR = '../chem_database'
Bohr2Ang = 0.52917721

def get_all_inputs():
    "get file names of all *.xyz input files"
    return sorted((os.path.splitext(filename)[0] for filename in os.listdir(INPUTS_DIR) if os.path.splitext(filename)[1] == '.xyz'))


x9_list = []
y9_list = []
for molecule in get_all_inputs():
    with open('{}/B3LYP/cc-pVQZ/VMC_OPT/emin/casl/8_8_44/10000/parameters.9.casl'.format(molecule), 'r') as f:
        casl = safe_load(f)

    jastrow = casl['JASTROW']

    try:
        for channel in ('Channel 1-1', 'Channel 1-2'):
            L = jastrow['TERM 1']['e-e cutoff']['Parameters'][channel]['L'][0]
            if channel == 'Channel 1-1':
                x9_list.append(L)
            else:
                y9_list.append(L)
            # print('{:14} {:14} {}'.format(molecule, channel, L * Bohr2Ang))
    except:
        print(molecule)
        
x5_list = []
y5_list = []
for molecule in get_all_inputs():
    with open('{}/B3LYP/cc-pVQZ/VMC_OPT/emin/casl/8_8_44/10000/parameters.5.casl'.format(molecule), 'r') as f:
        casl = safe_load(f)

    jastrow = casl['JASTROW']

    try:
        for channel in ('Channel 1-1', 'Channel 1-2'):
            L = jastrow['TERM 1']['e-e cutoff']['Parameters'][channel]['L'][0]
            if channel == 'Channel 1-1':
                x5_list.append(L)
            else:
                y5_list.append(L)
            # print('{:14} {:14} {}'.format(molecule, channel, L * Bohr2Ang))
    except:
        print(molecule)
```

```python
%matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt

plt.xlabel('cutoff for parallel spin')
plt.ylabel('cutoff for antiparallel spin')
plt.title('TERM 1 manipulation constant')
plt.grid(True)
plt.plot(x5_list, y5_list, 'go')
plt.plot(x9_list, y9_list, 'ro')
plt.show()
```
